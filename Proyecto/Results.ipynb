```python
import pandas as pd
data25=pd.read_csv('test_alpha25.csv',delimiter=',')
data5=pd.read_csv('test_alpha5.csv',delimiter=',')
data75=pd.read_csv('test_alpha75.csv',delimiter=',')
```

```python
data75.head()
```

```python
states=pd.concat([data25['Number_states'], data5['Number_states'],data75['Number_states']], 
                 axis=1, keys=['alpha=0.25','alpha=0.5','alpha=0.75'])
```

```python
states.head()
```

```python
rewards=pd.concat([data25['Accumulated_reward'], data5['Accumulated_reward'],data75['Accumulated_reward']],
                  axis=1, keys=['alpha=0.25','alpha=0.5','alpha=0.75'])
```

```python
rewards.head()
```

```python
times=pd.concat([data25['Time'], data5['Time'],data75['Time']],
                  axis=1, keys=['alpha=0.25','alpha=0.5','alpha=0.75'])
```

```python
times.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np


fig, ax = plt.subplots(figsize=(16,9))

ax.plot(states['alpha=0.25'],  label='alpha=0.25')
ax.plot(states['alpha=0.5'], label='alpha=0.5')
ax.plot(states['alpha=0.75'], label='alpha=0.75')

ax.legend(loc='best')
ax.set_ylabel('Number_states')
ax.set_xlabel('Trial')
ax.set_title('Number States per Trial')

plt.show
```

```python
fig, ax = plt.subplots(figsize=(16,9))

ax.plot(rewards['alpha=0.25'],  label='alpha=0.25')
ax.plot(rewards['alpha=0.5'], label='alpha=0.5')
ax.plot(rewards['alpha=0.75'], label='alpha=0.75')

ax.legend(loc='best')
ax.set_ylabel('Accumulated_reward')
ax.set_xlabel('Trial')
ax.set_title('Accumulated reward per Trial')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16,9))

ax.plot(times['alpha=0.25'],  label='alpha=0.25')
ax.plot(times['alpha=0.5'], label='alpha=0.5')
ax.plot(times['alpha=0.75'], label='alpha=0.75')

ax.legend(loc='best')
ax.set_ylabel('Time')
ax.set_xlabel('Trial')
ax.set_title('Time per Trial')

plt.show()
```
