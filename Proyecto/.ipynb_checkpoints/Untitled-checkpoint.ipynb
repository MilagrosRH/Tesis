```python
import pandas as pd
data=pd.read_csv('alimentos_data.csv', delimiter=',',encoding = 'utf8')
```

```python
data.head()
```

```python
data=data.filter(items=['nombre','grupo', 'energykcal'])
```

```python
datafilter=pd.read_csv('alimentos_lite.csv', delimiter=';')
datafilter
```

```python
import re
alimentos=[]
for i in datafilter[datafilter.columns[0]]:

    pattern = re.compile('^'+i+" ?")
    a=0;
    for j in data[data.columns[0]]:
        if pattern.match(j.lower()):
            alimentos.append([i,data.loc[a][1],data.loc[a][2]])
        a+=1
            
        
```

```python
#import numpy as np
#alimentos=np.array(alimentos)
alimentos
```

```python
data_Final=pd.DataFrame(alimentos)
data_Final
```

```python
data_Final.to_csv('data_Alimentos.csv')
```
