```python
class AgentRecommender():

    def __init__(self, program=None,learning_rate=0.01, reward_decay=0.9, e_greedy=0.9):
        self.alive = True
        self.lr = learning_rate
        self.gamma = reward_decay
        self.epsilon = e_greedy
        self.actions = []
        self.agents = []
        self.q_table = pd.DataFrame(columns=self.actions)
        if program is None or not isinstance(program, collections.Callable):
            print("Can't find a valid program for {}, falling back to default.".format(
                self.__class__.__name__))

            def program(percept):
                return eval(input('Percept={}; action? '.format(percept)))

        self.program = program
        

        
    def diet(self, evidence,inference):
        state=inference.likelihood_weighted_sample(evidence=evidence, size=1)
    
    def is_State_terminal(self,state):
        sumSymptoms=0
        for i in state:
            sumSymptoms+= i.level
        return (sumSymptoms==0)  
  
    def q_value(self,state,a):
        count =0
        for i in a:
            if(a.level==0):
                count+=1
        

    def learn(self, s, a, r, s_):
        q_predict = q_value(s, a)
        if not self.is_State_terminal(s_):
            q_target = r + self.gamma * q_value(s_, :)].max()  # next state is not terminal
        else:
            q_target = r  
        self.q_table.iloc[s, a] += self.lr * (q_target - q_predict)  # update
```
